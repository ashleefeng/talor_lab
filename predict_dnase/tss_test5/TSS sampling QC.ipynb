```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gm_close = "ENCFF342EGB_closed2tss.csv"
gm_open = "ENCFF342EGB_open2tss.csv"
k_close = "K562_rep1_closed2tss.csv"
k_open = "K562_rep1_open2tss.csv"
gm_rand_close = "ENCFF342EGB_closed2tss_random.csv"
gm_rand_open = "ENCFF342EGB_open2tss_random.csv"
k_rand_close = "K562_rep1_closed2tss_random.csv"
k_rand_open = "K562_rep1_open2tss_random.csv"
```

```python
ko = np.loadtxt(k_open)
kc = np.loadtxt(k_close)
krc = np.loadtxt(k_rand_close)
plt.figure(figsize=(10, 6))
plt.hist(kc, bins=100, range=(0, 150000), alpha=1 , histtype='step')
plt.hist(ko, bins=100, range=(0, 150000), alpha=1, histtype='step')
plt.hist(krc, bins=100, range=(0, 150000), alpha=1, histtype='step')
plt.xlabel('Distance to TSS (bp)')
plt.ylabel('Count')
plt.legend(["closed (tss)", "open", "closed (rand)"])
plt.show()
print("open mean: ", np.mean(ko))
print("closed mean: ", np.mean(kc))
print("random closed mean", np.mean(krc))
```
